```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
avg_rating = {}
df_all = pd.DataFrame()

def get_rating(path, movie_title, avg_rating):
  temp_df = pd.read_csv(path, index_col=0)
  temp_df["review date"] = pd.to_datetime(temp_df["review date"])
  before_nom = temp_df[temp_df["review date"] < datetime(2021, 3, 15)]
  before_nom = before_nom.dropna(axis=0)
  before_nom = before_nom[before_nom["rating"] != -1]
  before_nom_avg = before_nom["rating"].mean()
  after_nom = temp_df[temp_df["review date"] >= datetime(2021, 3, 15)]
  after_nom = after_nom.dropna(axis=0)
  after_nom = after_nom[after_nom["rating"] != -1]
  after_nom_avg = after_nom["rating"].mean()
  avg_rating[movie_title] = {"before_nom":before_nom_avg, "after_nom":after_nom_avg}
  temp_df = temp_df.assign(Title=movie_title)
  return temp_df
```

```python
title = []
before_rate = []
after_rate = []

for key in avg_rating.keys():
  title.append(key)
  after_rate.append(avg_rating[key]["after_nom"])
  before_rate.append(avg_rating[key]["before_nom"])

fig = plt.figure(figsize=(30,7))
len_x = np.arange(len(title)) * 2
plt.bar(len_x - 0.2, before_rate, 0.4, label='before nomination avg rate')
plt.bar(len_x + 0.2, after_rate, 0.4, label='after nomination avg rate')
plt.xlabel("Movie Title", fontsize=15)
plt.ylabel("Rating", fontsize=20)
plt.xticks(len_x, title, fontsize=7)
plt.title("Rating of oscar movies before and after nomination", fontsize=20)
plt.legend()
plt.show()
```
